# First we read the comments file and encode it to uft8
# Remove quotes

```python
list1=[]
filepath= 'comments.txt'
with open(filepath, encoding="utf8") as fp:  
   line = fp.readline()
   while line:
       list1.append(line.strip('"'))
       line = fp.readline()
```

# Convert file to Pandas DataFrame

import pandas as pd
df= pd.DataFrame(list1)

# Use vader sentiment analysis for polarity scores of 1st column

```python
df[0]
```

```python
from nltk.sentiment.vader import SentimentIntensityAnalyzer #Vader 
commentSentiment=[]

sid = SentimentIntensityAnalyzer()
for line in df[0]:
    ss = sid.polarity_scores(line)
    commentSentiment.append(ss)
commentSentiment
sentimentdf=pd.DataFrame(commentSentiment)    
sentimentdf.to_csv("sentiment_output2.csv")    
```

# Save comment sentiment to csv so we do not have to run previous steps on future iterations

```python
df2=df.join(sentimentdf)
df2.to_csv("Comments with Sentiment")
```
